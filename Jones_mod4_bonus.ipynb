```python
#Lee Jones Bonus - Mod4
print("Lee Jones")
path_to_file = 'C:/Users/lejones/Desktop/NWMO/CSIS-44608/608-mod4/Star_Spangled_Banner.txt'

word_counts = {}

with open(path_to_file) as text:
    for line in text:
        print(line.strip())
        for word in line.split():
            if word in word_counts:
                word_counts[word] += 1
            else:
                word_counts[word] = 1

print(f'{"WORD":<12}COUNT')

for word, count in sorted(word_counts.items()):
    print(f'{word:<12}{count}')

print('\nNumber of unique words:', len(word_counts))
```
